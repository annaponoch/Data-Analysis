```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# EDA

```python
lifedata = pd.read_csv("Life Expectancy Data.csv", header=0)
lifedata.head()
```

```python
lifedata.info()
```

```python
lifedata.isnull().sum()
```

```python
lifedata.tail(10)
```

```python
country_list = lifedata['Country'].unique()
metrics_list = lifedata.keys()
for country in country_list:
    lifedata.loc[lifedata['Country'] == country, metrics_list] = lifedata.loc[lifedata['Country'] == country,metrics_list].interpolate()
lifedata.dropna(inplace=True)
lifedata.isnull().sum()
```

е) Чи повинна країна з нижчим значенням очікуваної тривалості життя (<65) збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя?


```python
plt.figure(figsize=(4, 4))
dataset4=lifedata
bins = [0, 65, np.inf]
category = ['less 65', 'more 65']
dataset4['Period'] = pd.cut(dataset4['Life expectancy '], bins, labels=category)
dataset4= dataset4.dropna(subset=['Life expectancy ', 'Total expenditure'])
sns.scatterplot(data=dataset4, x="Life expectancy ", y="Total expenditure", hue="Period", style="Period")
```

З графіку можемо зробити висновок, що країна з нижчим значенням очікуваної тривалості життя (<65) необідно збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя
